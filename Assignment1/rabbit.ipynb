```python
from collections import deque
start = ['E', 'E', 'E', '_', 'W', 'W', 'W']
goal = ['W', 'W', 'W', '_', 'E', 'E', 'E']
def get_neighbors(state):
    neighbors = []
    pos = state.index('_')
    if pos > 0:
        if state[pos - 1] == 'E':
            new_state = state.copy()
            new_state[pos], new_state[pos - 1] = new_state[pos - 1], new_state[pos]
            neighbors.append(new_state)
    if pos > 1:
        if state[pos - 2] == 'E':
            new_state = state.copy()
            new_state[pos], new_state[pos - 2] = new_state[pos - 2], new_state[pos]
            neighbors.append(new_state)
    if pos < len(state) - 1:
        if state[pos + 1] == 'W':
            new_state = state.copy()
            new_state[pos], new_state[pos + 1] = new_state[pos + 1], new_state[pos]
            neighbors.append(new_state)
    if pos < len(state) - 2:
        if state[pos + 2] == 'W':
            new_state = state.copy()
            new_state[pos], new_state[pos + 2] = new_state[pos + 2], new_state[pos]
            neighbors.append(new_state)

    return neighbors

# BFS Algorithm
def bfs(start, goal):
    queue = deque([(start, [])])
    visited = set()

    while queue:
        state, path = queue.popleft()
        state_tuple = tuple(state)

        if state_tuple in visited:
            continue

        visited.add(state_tuple)

        if state == goal:
            return path + [state]

        for neighbor in get_neighbors(state):
            queue.append((neighbor, path + [state]))

    return None

# DFS Algorithm
def dfs(start, goal):
    stack = [(start, [])]
    visited = set()

    while stack:
        state, path = stack.pop()
        state_tuple = tuple(state)

        if state_tuple in visited:
            continue

        visited.add(state_tuple)

        if state == goal:
            return path + [state]

        for neighbor in get_neighbors(state):
            stack.append((neighbor, path + [state]))

    return None

# Run BFS and DFS
bfs_solution = bfs(start, goal)
dfs_solution = dfs(start, goal)
print("BFS Solution Path:")
if bfs_solution:
    for step in bfs_solution:
        print(' '.join(step))
else:
    print("No solution found.")
print("\nDFS Solution Path:")
if dfs_solution:
    for step in dfs_solution:
        print(' '.join(step))
else:
    print("No solution found.")
```
